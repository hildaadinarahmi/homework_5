## Lambda Function

```python
df
```

```python
def get_category(price):
  category = ''
  if price <= 500:
    category = 'Murah'
  else:
    category = 'Mahal'
  return category
```

```python
get_category(1400)
```

```python
for i in range(len(df)):
  df.loc[i, 'category'] = get_category(df.loc[i, 'Price ($)'])
```

```python
df
```

```python
df['category'].value_counts()
```

```python
df['Price ($)'].apply(lambda x: 'Murah' if x <= 500 else 'Mahal') # menentukan kategori hape pakai lambda
```
